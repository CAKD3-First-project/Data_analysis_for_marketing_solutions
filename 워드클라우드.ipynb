```python
import pandas as pd
single = pd.read_csv('/content/drive/MyDrive/cakd3_colab/my/dataset/1인가구_가공본.csv')
single.head()
```

```python
# 본문 중 2글자 이상 명사 추출 사용자 함수
from konlpy.tag import Okt
import re
def noun2_list(df):
  df_copy = df.copy()
  df_copy['body'] = df_copy['body'].apply(lambda x: re.sub('[^가-힣a-zA-Z]+', ' ', x))
  df_copy['noun'] = df_copy['body'].apply(lambda x : Okt().nouns(x))
  words = sum(df_copy['noun'].values.tolist(), [])
  noun_list = []
  for word in words:
    if len(word) > 1 :
      noun_list.append(word)
  return noun_list

single_noun_list = noun2_list(single)
single_noun_list
```

```python
# 불용어 제거
stopwords = ['가구', '증가', '만원', '기자', '이상', '경우', '전체', '매출', 
             '통해', '때문', '대한', '조사', '위해', '억원', '최근', '한국', '가장',
             '올해', '지난해', '내년', '관련', '비중', '분양', '제품', '대표',
             '배포', '관계자', '서울시', '금지', '사용', '지난', '시장', '소득',
             '이하', '최대', '업체', '메뉴', '대해', '대비', '가운데', '상품', 
             '신축', '개선', '관리', '뉴스', '비율', '정부', '반영', '역시', '발표',
             '이번', '기준', '통계청', '크게', '또한', '자료', '설명', '인기', '도입',
             '전체', '규모', '창업', '이후', '사진', '서비스', '서울', '단지', '정도',
             '응답', '기존', '라며', '다른', '가지', '수준', '차지', '모든', '더욱', 
             '한편', '연구원', '무단', '반면', '현재', '별로', '바로', '업계']

filtered_words = []
for word in single_noun_list:
  if word not in stopwords:
    filtered_words.append(word)


# counter 딕셔너리 사용자 함수
from collections import Counter
def most_common(list_name):
  count = Counter(list_name)
  words = dict(count.most_common())
  return words

single_words = most_common(filtered_words)
print(single_words)
```

```python
# 워드클라우드 생성
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (15,10))
wordcloud = WordCloud(font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf', background_color = 'white',
                      colormap = 'coolwarm', width = 1500, height = 1000).generate_from_frequencies(single_words) 

                      280 * 192
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
